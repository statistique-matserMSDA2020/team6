```python
"""
MATHIAM FAYE 
THILOR SENE
FATOU DIARRA
"""
import numpy.random as npr
import matplotlib.pyplot as plt
import numpy as np
nbre=10000                              #TAILLE DE L'ECHANTILLON DE LA LOI B(30,0.2)
t=npr.binomial(30,0.2,nbre)              
print(t)
p=[0]*30
for i in range(1,30):
    p[i]=t.tolist().count(i)/nbre    #CALCUL DE LA FREQUENCE D'APPARUTION DE CHAQUE RESULTAT
plt.hist(t,29,facecolor='green',range=(0.5,29.5))
plt.title("échantillon de taille 10000 suivant une loibinomiale B(30,0.2)".format(nbre))
plt.show()
```

```python
### Gamma distribution
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gamma
gamma_distribution = gamma(loc = 10, scale = 10, a = 5)
x = np.linspace(-5, 30, 10000)
_, ax = plt.subplots(1, 1)
ax.plot(x, gamma_distribution.pdf(x), 'r', lw=4)
print(x)
plt.title(' un échantillon de taille 10000 suivant une loi gamma γ(10,.5).')
plt.show()
print(gamma_distribution.pdf(x))
```
