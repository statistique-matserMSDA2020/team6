```python
"""
MATHIAM FAYE 
THILOR SENE
FATOU DIARRA
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Nombre de flechettes lanche dans le carr unite
N=10000

#la position(abcisse et ordonne) de l'impact choisie aleatoirement selon une loi uniforme 
x=np.random.uniform(low=0, high=1, size=[N,1])
y=np.random.uniform(low=0, high=1, size=[N,1])

#les points qui impacterons dans le cercle de rayon 1
plus=y**2<=1-x**2

#estimation de pi/4
pie_sur4=np.sum(plus)/N
print("VALEUR ESTIMEE DE PI/4:", pie_sur4,"POUR N=10000")
print("VRAIE VALEUR DE PI/4:", np.pi/4)

xplus=x[plus]
yplus=y[plus]
plt.figure(figsize=[5,5])
plt.scatter(x, y, s=6)
plt.scatter(xplus,yplus, color='r', s=6)
plt.show()
```
