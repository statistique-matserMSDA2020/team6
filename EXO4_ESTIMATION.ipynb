```python
"""
MATHIAM FAYE 
THILOR SENE
FATOU DIARRA
"""
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Nombre de flechettes lanchées dans le carré unite
iters=[10,100,1000,10000]
ity=np.array(iters)

for i in range(4):
    N=ity[i]
    print(N)
    #la position(abcisse et ordonnée) de l'impact choisie aleatoirement selon une loi uniforme  entre 0 et 1
    x=np.random.uniform(low=0, high=1, size=[N,1])
    y=np.random.uniform(low=0, high=1, size=[N,1])

    #les points qui impacterons dans le quart de cercle  de rayon 1
    plus=y<=np.sqrt(1-x**2)

    #estimation de pi/4
    pie_sur4=np.sum(plus)/N
    print("VALEUR ESTIMEE DE PI/4:", pie_sur4,"POUR N=",N)
    print("VRAIE VALEUR DE PI/4:", np.pi/4)

    xplus=x[plus]
    yplus=y[plus]
    plt.figure(figsize=[5,5])
    plt.scatter(x, y, s=6)
    plt.scatter(xplus,yplus, color='r', s=6)
    plt.show()
```
